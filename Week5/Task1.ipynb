```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv("WS4_1.csv")
df.head()
```

```python
num_cols_original = ["ASA", "a_ICM", "balabanJ", "b_count", "chi1", "dens", "E_sol", "E_str", "glob", "h_logS", "radius", "rgyr", "SlogP", "Weight", "weinerPath", "weinerPol", "zagreb"]
num_col_new = ["ASA", "a_ICM", "dens", "E_sol", "E_str", "glob", "h_logS", "radius"]
num_cols = num_col_new
method = 'pearson'
correlation_matrix = df[num_cols].corr(method=method)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(correlation_matrix, cmap='coolwarm', square=True, cbar_kws={'label': f'{method} Correlation Coefficient'})
plt.title('Correlation Matrix')
```

```python
method = 'kendall'
correlation_matrix = df[num_cols].corr(method=method)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(correlation_matrix, cmap='coolwarm', square=True, cbar_kws={'label': f'{method} Correlation Coefficient'})
plt.title('Correlation Matrix')
```

```python
method = 'spearman'
correlation_matrix = df[num_cols].corr(method=method)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(correlation_matrix, cmap='coolwarm', square=True, cbar_kws={'label': f'{method} Correlation Coefficient'})
plt.title('Correlation Matrix')
```
